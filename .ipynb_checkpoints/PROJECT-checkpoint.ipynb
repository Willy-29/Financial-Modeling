```python
dataSPXS=pd.read_csv('SPXS.csv',index_col='Date', parse_dates=True)#Loading data
dataSPXS
dataSPXL=pd.read_csv('SPXL.csv',index_col='Date', parse_dates=True)
dataSPXL
dataSPY=pd.read_csv('SPY.csv', index_col='Date', parse_dates=True)
dataSPY
```

```python
print(dataSPXS.isna().values.any())#checking that there is no NaNs isnull().values.any()
print(dataSPXL.isnull().values.any())
print(dataSPY.isnull().values.any())
```

```python
dataSPXS_close=dataSPXS.iloc[:,[3]]#Focussing on the close price
dataSPXL_close=dataSPXL.iloc[:,[3]]
dataSPY_close=dataSPY.iloc[:,[3]]
dataSPXS_close
```

```python
fig, ax = plt.subplots(1, 1)#plotting historical close prices
ax.plot(dataSPXS_close, label ='SPXS Close Price')
ax.plot(dataSPXL_close, label ='SPXL Close Price')
ax.plot(dataSPY_close, label ='SPY Close Price')

ax.set_xlabel('Date',fontsize=15)
ax.tick_params(axis='x', rotation=45,)
ax.set_ylabel('Close price',fontsize=15)
ax.legend()
plt.title("Close Prices of the 3 ETFs since 2017")
ax.grid(axis='x', linestyle='--')
ax.grid(axis='y', linestyle='--')
fig.tight_layout() 
plt.show()
```

```python
dataSPXS_close_log_ret=np.log(dataSPXS_close).pct_change()#computing the daily log-returns of the close prices
dataSPXS_close_log_ret.rename(columns={dataSPXS_close_log_ret.columns[0]:'SPXS Log-Returns'},inplace=True)
dataSPXS_close_log_ret.dropna(inplace=True)

dataSPXL_close_log_ret=np.log(dataSPXL_close).pct_change()
dataSPXL_close_log_ret.rename(columns={dataSPXL_close_log_ret.columns[0]:'SPXL Log-Returns'},inplace=True)
dataSPXL_close_log_ret.dropna(inplace=True)

dataSPY_close_log_ret=np.log(dataSPY_close).pct_change()
dataSPY_close_log_ret.rename(columns={dataSPY_close_log_ret.columns[0]:'SPY Log-Returns'},inplace=True)
dataSPY_close_log_ret.dropna(inplace=True)

data_log_returns=pd.concat([dataSPXS_close_log_ret,dataSPXL_close_log_ret,dataSPY_close_log_ret],axis=1)
data_log_returns



```

```python
fig, ax = plt.subplots(1, 1) #Plotting the returns (not very clear if plotted together)
#ax.plot(dataSPXS_close_log_ret, label ='SPXS Log Returns')
ax.plot(dataSPXL_close_log_ret, label ='SPXL Log Returns')
#ax.plot(dataSPY_close_log_ret, label ='SPY Log Returns')

ax.set_xlabel('Date',fontsize=15)
ax.tick_params(axis='x', rotation=45)
ax.set_ylabel('Log Returns',fontsize=15)
ax.legend()
plt.title("Log Returns of the 3 ETFs since 2017",fontsize=20)
ax.grid(axis='x', linestyle='--')
fig.tight_layout() 
plt.show()
```

## Power profiles and Gamma effects

```python
def period_return(per_ret):
    return np.prod(per_ret + 1)- 1
```

```python
rolling_returns=pd.DataFrame()
per_ret_SPXS=dataSPXS_close.pct_change()
rolling_SPXS= per_ret_SPXS.rolling('365D').apply(period_return)
rolling_returns['1 Year Rolling Returns SPXS']=rolling_SPXS.mul(100)

per_ret_SPXL=dataSPXL_close.pct_change()
rolling_SPXL= per_ret_SPXL.rolling('365D').apply(period_return)
rolling_returns['1 Year Rolling Returns SPXL']=rolling_SPXL.mul(100)

per_ret_SPY=dataSPY_close.pct_change()
rolling_SPY= per_ret_SPY.rolling('365D').apply(period_return)
rolling_returns['1 Year Rolling Returns SPY']=rolling_SPY.mul(100)

rolling_returns = rolling_returns.dropna()
rolling_returns
    
```

```python
plt.scatter(x=rolling_returns['1 Year Rolling Returns SPY'],y=rolling_returns['1 Year Rolling Returns SPXS'],s=1.)
plt.scatter(x=rolling_returns['1 Year Rolling Returns SPY'],y=rolling_returns['1 Year Rolling Returns SPXL'],s=1.)
plt.title('Historical plot of VT/V0',fontsize=10)
plt.show
```

```python
roll_ret_SPXS=[]
roll_ret_SPXL=[]
roll_ret_SPY=[]
for i in range(1319):
    roll_ret_SPXS.append((dataSPY_close.iloc[i+252,0]/dataSPY_close.iloc[i,0])**-3)
    roll_ret_SPXL.append((dataSPY_close.iloc[i+252,0]/dataSPY_close.iloc[i,0])**3)
    roll_ret_SPY.append(dataSPY.iloc[i+252,0]/dataSPY.iloc[i,0])
    

plt.scatter(x=roll_ret_SPY,y=roll_ret_SPXS,s=1.)    
plt.scatter(x=roll_ret_SPY,y=roll_ret_SPXL,s=1.)

plt.title("Plot of VT/V0 without trading impact",fontsize=10)
plt.show
```

### Gamma Effect

```python
def gamma_effect(l):
    return -0.5*(l**2-l)
rolling_returns['accumulated_var']=(data_log_returns['SPY Log-Returns']**2).rolling(252).sum().dropna()
rolling_returns['with_trade_impact_SPXS']=np.exp(gamma_effect(-3)*rolling_returns['accumulated_var'])*rolling_returns['1 Year Rolling Returns SPXS']
rolling_returns['with_trade_impact_SPXL']=np.exp(gamma_effect(3)*rolling_returns['accumulated_var'])*rolling_returns['1 Year Rolling Returns SPXL']
```

```python
plt.scatter(x=rolling_returns['1 Year Rolling Returns SPY'],y=rolling_returns['with_trade_impact_SPXS'],s=1.)    
plt.scatter(x=rolling_returns['1 Year Rolling Returns SPY'],y=rolling_returns['with_trade_impact_SPXL'],s=1.)
plt.title('Plot of VT/V0 with trading impact',fontsize=10)
```
